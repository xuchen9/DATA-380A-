```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s=0.2
δ=0.1
α=1/3
```

```python
def F(A,K,L):
    return A*K**α*L**(1-α)
```

```python
n=0.02
A=1
k=np.linspace(0,40,20)
```

1.1

```python
T=40
L=np.zeros(T)
K=np.zeros(T)
L[0]=1
K[0]=0.05


Y=np.zeros(T)
Y[0]=F(A,K[0],L[0])


s2=0.3
```

```python
#aggregate GDP

for t in range(1,T):
    
    if t<=20:
        L[t]=L[t-1]*(1+n) 
        K[t]=K[t-1]+s*Y[t-1]-δ*K[t-1]
        Y[t]=F(1,K[t],L[t])
    else:
        L[t]=L[t-1]*(1+n) 
        K[t]=K[t-1]+s2*Y[t-1]-δ*K[t-1]
        Y[t]=F(1,K[t],L[t])
```

```python
fig,ax=plt.subplots()
ax.plot(range(40), Y, label='aggregated GDP')
ax.legend()
```

```python
#GDP per person

fig,ax=plt.subplots()
ax.plot(range(T), Y/L, label='GDP per person')
ax.legend()
```

```python

a=63
γ=21/20
b=0
β=1

T=50
p=np.zeros(T)
p[0]=0

Pe=np.zeros(T) 
ε=np.zeros(T)

for t in range(1,T):
    Pe[t]=(a-b)/γ -(β/γ)*Pe[t]


for t in range(1,T):
    p[t]=60- (20/21)*(Pe[t]-60)-ε[t]
```

```python
fig,ax=plt.subplots()
ax.plot(range(T), p, label='positive feedback: RE price')
ax.legend()
```

```python
for t in range(1,T):
    p[t]=60+(20/21)*(Pe[t]-60)-ε[t]
```

```python
fig,ax=plt.subplots()
ax.plot(range(T), p, label='negative feedback: RE price')
ax.legend()
```
